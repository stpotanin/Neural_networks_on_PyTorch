```python
# в каждом списке 4 значения - количество блоков (по порядку)
# в слоях Layer1, Layer2, Layer3 и Layer4
from torchvision import models

model_18 = models.resnet18()
model_34 = models.resnet34()
model_50 = models.resnet50()

resnet18 = [len(model_18.layer1), len(model_18.layer2), len(model_18.layer3), len(model_18.layer4)]
resnet34 = [len(model_34.layer1), len(model_34.layer2), len(model_34.layer3), len(model_34.layer4)]
resnet50 = [len(model_50.layer1), len(model_50.layer2), len(model_50.layer3), len(model_50.layer4)]

print(resnet18)
print(resnet34)
print(resnet50)
```
