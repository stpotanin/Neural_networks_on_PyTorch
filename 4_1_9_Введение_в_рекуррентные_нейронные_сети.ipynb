```python
import torch
import torch.nn as nn

N = 200  #   
r = 0.95 #  
std = 10 #   sigma
std_e = std * (1 - r * r) ** 0.5 #    

x = torch.empty(N, dtype=torch.float32)
x[0] = torch.randn(1) * std                    # 0 + _t

for t in range(1, len(x)):
    x[t] = r * x[t-1] + torch.randn(1) * std_e # x_t = r * x_{t-1} + _t

# # 
# import matplotlib.pyplot as plt

# xx = range(len(x))
# plt.plot(xx, x)
# plt.title(f'r = {r}')
# plt.show()
```
