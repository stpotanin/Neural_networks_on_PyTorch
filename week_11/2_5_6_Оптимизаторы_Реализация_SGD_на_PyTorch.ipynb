```python
import torch

# Функции
func = lambda x: 0.2 * (x - 2) ** 2 - 0.3 * torch.cos(4 * x)

# Параметры для ГС
lr = 0.1    # шаг обучения
x0 = 0.0    # начальное значение точки минимума
N = 200     # число итераций градиентного алгоритма
x = torch.tensor(x0, requires_grad=True)
optimizer = torch.optim.RMSprop(params=[x], lr=lr)

# ГС
for _ in range(N):
    y = func(x)
    y.backward()
    optimizer.step()        # шаг с заданным выше оптимизатором
    optimizer.zero_grad()   # обнулениие накопленного гадиента

# print(f'Минимум в точке {x}')

# График
import matplotlib.pyplot as plt

x_plot = torch.linspace(-1.5, 5.0, 100)
y_plot = func(x_plot)
x_min = x.detach().numpy()
y_min = func(x).detach().numpy()
plt.plot(x_plot, y_plot)
plt.scatter(x_min, y_min, c='r')
plt.grid()
plt.title('Поиск глобального минимума с использованием оптимизатора ГС')
plt.show()
```
