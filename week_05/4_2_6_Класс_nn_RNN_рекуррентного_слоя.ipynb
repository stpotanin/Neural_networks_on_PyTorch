```python
import torch
import torch.nn as nn

class GetOutput(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        output, hidden_state = x
        # output.shape: (batch_size, sequence_length, hidden_size)
        # hidden_state.shape: (num_layers * num_directions, batch_size, hidden_size)
        # т.к. один слой (num_layers=1) и не указан параметр bidirectional,
        # то num_directions=1, и форма будет (1, batch_size, hidden_size)
        # print(hidden_state.shape)
        return hidden_state

# тензор x в программе не менять
batch_size = 4
seq_length = 8
in_features = 10
x = torch.rand(batch_size, seq_length, in_features)

model = nn.Sequential(
    nn.RNN(in_features, 15, batch_first=True),
    GetOutput(),
    nn.ReLU(inplace=True),
    nn.Linear(15, 5)
)

model.eval()
res = model(x)

res.shape
```

```python
# @title Эталонное решение https://github.com/selfedu-rus/neuro-pytorch/blob/main/solves/4.2.4
import torch
import torch.nn as nn

# здесь объявляйте класс GetOutput
class GetOutput(nn.Module):
    def forward(self, x):
        return x[1].squeeze(0)


# тензор x в программе не менять
batch_size = 4
seq_length = 8
in_features = 10
x = torch.rand(batch_size, seq_length, in_features)

# здесь продолжайте программу
model = nn.Sequential(
    nn.RNN(in_features, 15, batch_first=True),
    GetOutput(),
    nn.ReLU(inplace=True),
    nn.Linear(15, 5)
)

model.eval()
res = model(x)

res.shape
```
