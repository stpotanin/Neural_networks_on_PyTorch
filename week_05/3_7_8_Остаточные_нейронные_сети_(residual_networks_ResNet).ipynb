```python
import torch
import torch.nn as nn

class Block(nn.Module):
    def __init__(self):
        super().__init__()
        self.block = nn.Sequential(
            # Вход: (batch_size, 64, 32, 32)
            nn.Conv2d(64, 64, (3, 3), stride=1, padding=1, bias=False),  # (batch_size, 64, 32, 32)
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, (3, 3), stride=1, padding=1, bias=False),  # (batch_size, 64, 32, 32)
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        x = x + self.block(x)
        return x

batch_size = 8
x = torch.rand(batch_size, 64, 32, 32) # тензор x в программе не менять

model_bb = Block()
model_bb.eval()

y = model_bb(x)

# print(y.shape)
```
