```python
import torch
import torch.nn as nn

class My_Class(nn.Module):
    def __init__(self):
        super().__init__()
        self.inp_1 = nn.Sequential(
            nn.Linear(7, 12),
            nn.Tanh()
        )
        self.inp_2 = nn.Sequential(
            nn.Linear(12, 12),
            nn.Sigmoid()
        )
        self.out = nn.Sequential(
            nn.Linear(12, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, a, b):
        return self.out(self.inp_1(a) + self.inp_2(b))

batch_size=12
a = torch.rand(batch_size, 7) # тензоры a, b в программе не менять
b = torch.rand(batch_size, 12)

model = My_Class()
model.eval()

predict = model(a, b)
predict.shape
```

```python
import torch
import torch.nn as nn

class ModelNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.inp_1 = nn.Sequential(
            nn.Linear(7, 12),
            nn.Tanh()
        )
        self.inp_2 = nn.Sequential(
            nn.Linear(12, 12),
            nn.Sigmoid()
        )
        self.out = nn.Sequential(
            nn.Linear(12, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x1, x2):
            x = self.inp_1(x1) + self.inp_2(x2)
            x = self.out(x)
            return x

# Данные
batch_size=12
a = torch.rand(batch_size, 7)
b = torch.rand(batch_size, 12)

model = ModelNN()
model.eval()

predict = model(a, b)

# Проверка
print(predict.shape)
for name, param in model.named_parameters():
    print(f"Имя параметра: {name}")
```
