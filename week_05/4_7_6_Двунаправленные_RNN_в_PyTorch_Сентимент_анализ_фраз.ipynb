```python
import torch
import torch.nn as nn

class One_to_Many(nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn = nn.RNN(d_size, hidden_size, batch_first=True, bidirectional=True)
        self.out = nn.Linear(hidden_size * 2, 2)

    def forward(self, x):
        x = x.unsqueeze(1)
        x = torch.cat([x, torch.zeros(batch_size, 11, d_size)], dim=1)
        rnn_output = self.rnn(x)[0]  # извлекаем только выходы RNN
        return self.out(rnn_output)

# тензор x в программе не менять
batch_size = 4
seq_length = 12
d_size = 5
x = torch.rand(batch_size, d_size)
x.shape

in_linear = 14               # из рисунка
out_size = 2                 # из рисунка
hidden_size = in_linear // 2 # т.к. bidirectional

model = One_to_Many()

model.eval()
predict = model(x)
predict.shape   # [4, 12, 2]
```
