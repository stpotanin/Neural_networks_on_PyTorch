```python
import torch
import torch.nn as nn

class BasicBlock1(nn.Module):
    def __init__(self):
        super().__init__()
        # Вход (batch_size, 64, H, W)
        self.bb_1 = nn.Sequential(
            nn.Conv2d(64, 128, (3, 3), stride=2, padding=1, bias=False), # (batch_size, 128, H/2, W/2)
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, (3, 3), stride=1, padding=1, bias=False), # (batch_size, 128, H/2, W/2)
            nn.BatchNorm2d(128)
        )

        self.bb_2 = nn.Sequential(
            nn.Conv2d(64, 128, (1, 1), stride=2, padding=0, bias=False), # (batch_size, 128, H/2, W/2))
            nn.BatchNorm2d(128)
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        return self.relu(self.bb_1(x) + self.bb_2(x))                    # (batch_size, 128, H/2, W/2)

class BasicBlock2(nn.Module):
    def __init__(self):
        super().__init__()
        # Вход (batch_size, 128, H/2, W/2)
        self.bb = nn.Sequential(
            nn.Conv2d(128, 128, (3, 3), stride=1, padding=1, bias=False), # (batch_size, 128, H/2, W/2)
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.Conv2d(128, 128, (3, 3), stride=1, padding=1, bias=False), # (batch_size, 128, H/2, W/2)
            nn.BatchNorm2d(128)
        )

        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        return self.relu(self.bb(x) + x)                                  # (batch_size, 128, H/2, W/2)

batch_size = 8
x = torch.rand(batch_size, 3, 32, 32) # тензор x в программе не менять

model = nn.Sequential(
            nn.Conv2d(3, 64, (7, 7), stride=2, padding=3, bias=False),  # (batch_size, 64, 16, 16)
            nn.BatchNorm2d(64),
            nn.MaxPool2d((3, 3), stride=2, padding=1),                 # (batch_size, 64, 8, 8)
            BasicBlock1(),                                             # (batch_size, 128, 4, 4)
            BasicBlock2(),                                             # (batch_size, 128, 4, 4)
            nn.AdaptiveAvgPool2d(1),                                   # (batch_size, 128, 1, 1)
            nn.Flatten(),                                              # (batch_size, 128)
            nn.Linear(128, 10, bias=True)                              # (batch_size, 10)
)

model.eval()
predict = model(x)

# predict.shape
```
