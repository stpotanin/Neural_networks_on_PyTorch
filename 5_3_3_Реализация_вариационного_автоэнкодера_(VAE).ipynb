```python
import torch
import torch.nn as nn

# здесь объявляйте класс VAE
class VAELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_1 = nn.Linear(128, 32,bias=False)
        self.batch_norm_1 = nn.BatchNorm1d(32)
        self.linear_2 = nn.Linear(32, 8,bias=False)
        self.batch_norm_2 = nn.BatchNorm1d(8)
        self.linear_3 = nn.Linear(8, 4)
        self.linear_4 = nn.Linear(8, 4)
        self.linear_5 = nn.Linear(4, 64)
        self.linear_6 = nn.Linear(64, 128)

    def forward(self, x):
        # encoder
        x = self.batch_norm_1(self.linear_1(x)).relu()
        x = self.batch_norm_2(self.linear_2(x)).relu()
        h_mean = self.linear_3(x)
        h_var = self.linear_4(x).relu()
        noise = torch.normal(mean=torch.zeros_like(h_mean), std=torch.ones_like(h_var))
        h = h_var**.5 * noise + h_mean

        # decoder
        y = self.linear_5(h).relu()
        y = self.linear_6(y)

        return y, h_mean, h_var

# тензор data_x в программе не менять
batch_size = 8
data_x = torch.rand(batch_size, 128)

# здесь создавайте модель и продолжайте программу
model = VAELoss()
model.eval()
out, hm, hv = model(data_x)
out.shape, hm.shape, hv.shape
```
