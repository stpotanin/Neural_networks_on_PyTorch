```python
import torch
import torch.nn as nn
from collections import OrderedDict

model = nn.Sequential(OrderedDict([
          ('layer1', nn.Linear(12, 24)),
          ('act1', nn.Tanh()),
          ('layer2', nn.Linear(24, 10)),
          ('act2', nn.Tanh()),
          ('out', nn.Linear(10, 1))
        ]))

batch_size=16
x = torch.rand(batch_size, 12) # тензор x в программе не менять

model.eval()
predict = model(x)
predict
```
