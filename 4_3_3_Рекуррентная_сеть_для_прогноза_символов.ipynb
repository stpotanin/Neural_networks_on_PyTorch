```python
import torch
import torch.nn as nn

sigma = 0.1 # стандартное отклонение отсчетов последовательности
r = 0.9 # коэффициент регрессии
sigma_noise = sigma * (1 - r * r) ** 0.5 # стандартное отклонение случайных величин

total = 100 # длина генерируемой последовательности
noise = torch.randn((total, )) # случайные величины, подаваемые на вход модели
x0 = torch.randn((1, )) * sigma # начальное значение вектора скрытого состояния

model = nn.RNN(input_size=1, hidden_size=1, bias=False, batch_first=True)
model.weight_hh_l0.data = torch.Tensor([r]).unsqueeze(0)
model.weight_ih_l0.data = torch.Tensor([sigma_noise]).unsqueeze(0)
model.eval()

# Размерность входных данных для RNN с batch_first=True: (batch_size, seq_len, input_size)
x, _ = model(noise.unsqueeze(0).unsqueeze(-1),  # (100,) -> (1,100,1)
             x0.unsqueeze(0).unsqueeze(0)  # (num_layers=1, batch_size=1, hidden_size=1)
             )

x.shape
```
