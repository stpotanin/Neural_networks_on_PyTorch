```python
import torch

class ConvTranspose2d_mini():
    def __init__(self) -> None:
        pass

    def __call__(self, x, kernel):
        kh, kw = kernel.shape[0], kernel.shape[1]
        H = x.shape[0] * kh
        W = x.shape[1] * kw
        out = torch.zeros(H, W, dtype=torch.int32)
        for i in range(x.shape[0]):
            for j in range(x.shape[1]):
                for m in range(kh):
                    for n in range(kw):
                        out[i * kh + m, j * kw + n] += x[i, j] * kernel[m, n]
        return out

kernel = torch.tensor([[-1, 2], [1, -2]], dtype=torch.int32)
x = torch.tensor([[1, 1], [2, 2]], dtype=torch.int32)
result = ConvTranspose2d_mini()(x, kernel)

print(result)
```
