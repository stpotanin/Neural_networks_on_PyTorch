```python
import torch

class ConvTranspose2d_midi():
    def __init__(self) -> None:
        pass

    def __call__(self, x, kernel, stride=1):
        kh, kw = kernel.shape[0], kernel.shape[1]
        H = (x.shape[0] - 1) * stride + kh
        W = (x.shape[1] - 1) * stride + kw
        out = torch.zeros(H, W, dtype=torch.int32)
        for i in range(x.shape[0]):
            for j in range(x.shape[1]):
                for m in range(kh):
                    for n in range(kw):
                        out[i * stride + m, j * stride + n] += x[i, j] * kernel[m, n] # учёт страйда
        return out


kernel = torch.tensor([[-1, 2, 3],
                       [1, -2, 1],
                       [-3, 1, 4]],
                      dtype=torch.int32)
x = torch.tensor([[1, 2], [-3, 4]], dtype=torch.int32)
stride = 2

result = ConvTranspose2d_midi()(x, kernel, stride)

# print(result)
```
