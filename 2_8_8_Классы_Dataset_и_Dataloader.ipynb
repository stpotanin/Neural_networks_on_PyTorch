```python
import torch
import torch.nn as nn
import torch.utils.data as data

# Функция f(x)=2⋅exp(−x/2)+0.2⋅sin(x/10)−5
f = lambda x: 2 * torch.exp(- x / 2) + 0.2 * torch.sin(x / 10) - 5

class FuncDataset(data.Dataset):
    def __init__(self):
        self.data = torch.arange(-5, 5, 0.1)
        self.targets = f(self.data)
        self.length = self.data.shape[0]

    def __getitem__(self, item):
        return self.data[item], self.targets[item]

    def __len__(self):
        return self.length

d_train = FuncDataset()
x13, y13 = d_train[13]

total = len(d_train)

# print(x13, y13)

# График
import torch.nn as nn
import matplotlib.pyplot as plt

plt.scatter(d_train.data.numpy(), d_train.targets.numpy(), s=10)
plt.scatter(x13.numpy(), y13.numpy(), c='r')
plt.grid()
plt.show()
```
