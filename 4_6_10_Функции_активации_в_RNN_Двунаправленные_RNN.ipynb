```python
import torch
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(5, 16, bias=False),
    nn.ReLU(),
    nn.BatchNorm1d(16),
    nn.Linear(16, 3)
)

batch_size=8
x = torch.rand(batch_size, 5) # тензор x в программе не менять

model.eval()
predict = model(x)
predict
```
