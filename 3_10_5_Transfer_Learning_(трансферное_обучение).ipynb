```python
import torch
from torchvision import models
import torchvision.transforms.functional as TF

#  x  img_pil    
x = torch.randint(0, 255, (3, 100, 100), dtype=torch.float32)
img_pil = TF.to_pil_image(x)
# display(img_pil.size, img_pil) #   

weights = models.ResNet34_Weights.DEFAULT
transforms = weights.transforms()
inp_img = transforms(img_pil)

# print(inp_img.shape)

model = models.resnet34()
model.requires_grad_(False)
model.eval()

results = model(inp_img.unsqueeze(0)) #   

# res = results.softmax(dim=1).sort(descending=True)
# cats = weights.meta['categories']
# for s, i in zip(res[0][0][:5], res[1][0][:5]):
#     print(f"{cats[i]}: {s:.4f}")
```
