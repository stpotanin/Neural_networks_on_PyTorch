```python
import torch
import torch.nn as nn

class DecoderBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.transpose = nn.ConvTranspose2d(in_channels, 16, kernel_size=2, stride=2, bias=True)
        self.block = nn.Sequential(
            nn.Conv2d(32, 16, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(16),
            nn.ReLU(inplace=True),
            nn.Conv2d(16, 16, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(16),
            nn.ReLU(inplace=True)
        )
        self.out = nn.Conv2d(16, out_channels, kernel_size=1, stride=1, padding=0)



    def forward(self, x, y):
        x = self.transpose(x)           # (batch_size, 16, 64, 64)
        u = torch.cat([x, y], dim=1)    # (batch_size, 32, 64, 64)
        u = self.block(u)               # (batch_size, 16, 64, 64)
        return self.out(u)              # (batch_size, out_channels, 64, 64)

# тензоры x, y в программе не менять
batch_size = 2
x = torch.rand(batch_size, 32, 32, 32)
y = torch.rand(batch_size, 16, 64, 64)

model = DecoderBlock(32, 1)
model.eval()
with torch.no_grad():
    out = model(x, y)

# print(out.shape)  # [2, 1, 64, 64]
```
