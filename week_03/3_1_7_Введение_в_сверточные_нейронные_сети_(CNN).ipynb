```python
import torch

# Функция
def pad(x, padding):
    # Строки нулей (C, 1, W)
    zero_rows = torch.zeros((x.shape[0], padding, x.shape[2]), dtype=x.dtype)
    x = torch.cat((zero_rows, x, zero_rows), dim=1)  # Добавляем сверху и снизу

    # Столбцы нулей (C, H + 2*padding, 1) - x.shape уже поменялся
    zero_cols = torch.zeros((x.shape[0], x.shape[1], padding), dtype=x.dtype)
    x = torch.cat((zero_cols, x, zero_cols), dim=2)  # Добавляем слева и справа

    return x

C = 2                   # число каналов
H, W = 16, 12           # размеры изображения: H - число строк; W - число столбцов
kernel_size = (5, 3)    # размер ядра по осям (H, W)
stride = (1, 1)         # шаг смещения ядра по осям (H, W)
padding = 1             # размер нулевой области вокруг изображения (число строк и столбцов с каждой стороны)

H_out = int((H + 2 * padding - kernel_size[0]) / stride[0] + 1)
W_out = int((W + 2 * padding - kernel_size[1]) / stride[1] + 1)

x_img = torch.randint(0, 255, (C, H, W), dtype=torch.float32) # тензоры x_img и kernel
kernel = torch.rand((C, ) + kernel_size) # в программе не менять

# Сначала паддинг входного тензора
x_img_padded = pad(x_img, padding)

# Теперь свёртка
predict = torch.empty(H_out, W_out, dtype=torch.float32)
for i in range(0, H_out):
    for j in range(0, W_out):
        predict[i, j] = torch.sum(x_img_padded[:, # по двум каналам
                                        i*stride[0]:kernel_size[0]+i*stride[0],
                                        j*stride[1]:kernel_size[1]+j*stride[1]] * kernel, dim=(0, 1, 2))

x_img.shape, predict.shape, x_img_padded, predict
```
