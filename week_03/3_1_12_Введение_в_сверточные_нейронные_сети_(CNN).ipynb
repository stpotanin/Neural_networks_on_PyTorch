```python
import torch

# Параметры
H, W = 24, 24           # размеры карты признаков: H - число строк; W - число столбцов
kernel_size = (3, 2)    # размер окна для Pooling по осям (H, W)
stride = (2, 2)         # шаг смещения окна по осям (H, W)
padding = 1             # размер нулевой области вокруг карты признаков (число строк и столбцов с каждой стороны)

# Данные (случайное изобаражение)
x = torch.rand((H, W)) # карта признаков (в программе не менять)

def AveragePooling(x, kernel, stride, padding=0):
    # Паддинг
    if padding:
        # Матрица из нулей, увеличенная на размер паддинга
        rezult = torch.zeros((x.shape[0] + 2 * padding, x.shape[1] + 2 * padding))
        # Заполнение "середины" исходным изображением
        rezult[padding:x.shape[0] + padding, padding:x.shape[1] + padding] = x
        x = rezult

    # Размеры выхода
    H_out = (x.shape[0] - kernel[0]) // stride[0] + 1
    W_out = (x.shape[1] - kernel[1]) // stride[1] + 1

    # AveragePooling непосредственно
    out = torch.empty(H_out, W_out, dtype=x.dtype) # заготовка по размерам
    for i in range(H_out):
        for j in range(W_out):
            # начальные координаты окна
            h_start = i * stride[0]
            w_start = j * stride[1]
            # окошко
            window = x[h_start:h_start + kernel[0], # окно размером
                       w_start:w_start + kernel[1]] # с ядро
            # среднее в окошке - находим и перезаписываем ячейку в выходе
            out[i, j] = window.mean()

    return out

res_pool = AveragePooling(x, kernel_size, stride, padding)

# print(res_pool)
```
