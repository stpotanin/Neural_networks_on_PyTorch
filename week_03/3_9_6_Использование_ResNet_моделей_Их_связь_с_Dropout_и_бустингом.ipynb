```python
import torch
from torchvision import models
import torchvision.transforms.functional as TF

# тензор x и img_pil в программе не менять
x = torch.randint(0, 255, (3, 128, 128), dtype=torch.float32)
img_pil = TF.to_pil_image(x)
# display(img_pil) # можно глянуть на картинку

# Загружеаем и трансформации для картинки
transforms = models.ResNet18_Weights.DEFAULT.transforms()

# Трансформируем картинку
inp_img = transforms(img_pil)

# Пропускаем тензор img через ResNet18 со случайными весами:
model = models.resnet18()
model.eval()
results = model(inp_img.unsqueeze(0)) # (1000) добавляем первую ось

# Применение Softmax, чтобы преобразовать значения
# в вероятности (сумма всех вероятностей равна 1);
# и сортировка по убыванию, чтобы получить индексы наиболее вероятных классов
res = results.softmax(dim=1).sort(descending=True)
resnet_weights = models.ResNet18_Weights.DEFAULT # дефолтные веса для названий классов
cats = resnet_weights.meta['categories']
for i in res[1][0][:4]:
    print(cats[i])
```
