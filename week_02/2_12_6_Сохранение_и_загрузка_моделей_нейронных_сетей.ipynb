```python
import torch
import torch.nn as nn

class ThreeLayerModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(13, 32)
        self.layer2 = nn.Linear(32, 16)
        self.layer3 = nn.Linear(16, 3)

    def forward(self, x):
        x = self.layer1(x).relu()
        x = self.layer2(x).relu()
        x = self.layer3(x)
        return x

# Модель
model = ThreeLayerModel()

# Сохраниние весов в перменную
data_w = model.state_dict()

# Сохранение в файл
torch.save(data_w, 'func_nn.tar')

# Проверка
data_w2 = torch.load('func_nn.tar', weights_only=True)
print(data_w2)
```
