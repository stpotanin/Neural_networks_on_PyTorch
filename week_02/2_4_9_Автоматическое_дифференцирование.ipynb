````python
import torch

# x0, x1, x2, x3 = map(float, input().split()) # переменные x0, x1, x2, x3 в программе не менять
# x0, x1, x2, x3 = 0.12, 0.5, -2.3, 0.1   # Тест 1
x0, x1, x2, x3 = 5.12, -1.5, 1.3, -0.5  # Тест 2

coords_x = torch.arange(-4, 6, 0.1) # точки интервала [-4; 6) с шагом 0.1 (тензор в программе не менять)

w0 = torch.tensor(x0, dtype=torch.float32, requires_grad=True)
w1 = torch.tensor(x1, dtype=torch.float32, requires_grad=True)
w2 = torch.tensor(x2, dtype=torch.float32, requires_grad=True)
w3 = torch.tensor(x3, dtype=torch.float32, requires_grad=True)

N = len(coords_x)
loss = 0

for x in coords_x:
    f = -0.7 * x - 0.2 * x**2 + 0.05 * x**3 - 0.2 * torch.cos(3 * x) + 2
    a = w0 + w1 * x + w2 * x**2 + w3 * x**3
    loss += (a - f) ** 2 / N

loss.backward()

'''График'''
import matplotlib.pyplot as plt
import numpy as np

# Исходная функция
f_func = lambda x: -0.7 * x - 0.2 * x**2 + 0.005 * x**3 - 0.2 * torch.cos(3 * x) + 2

# Аппроксимирующая функция
a_func = lambda x: w0 + w1 * x + w2 * x**2 + w3 * x**3

# Вычисление функций в точках интервала
y_f = f_func(coords_x)
y_a = a_func(coords_x)

# Отключение тензоров от вычислительного графа и преобразование в NumPy
y_f_np = y_f.detach().numpy()
y_a_np = y_a.detach().numpy()
coords_x_np = coords_x.detach().numpy()

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(coords_x_np, y_f_np, label='Исходная функция')
plt.plot(coords_x_np, y_a_np, label='Аппроксимирующая функция')
plt.legend()
plt.title('График функций Тест 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
````
