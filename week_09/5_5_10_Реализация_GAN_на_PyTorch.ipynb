```python
import torch
import torch.nn as nn

class Bottleneck(nn.Module):
    def __init__(self):
        super().__init__()
        self.direct = nn.Sequential(
            nn.Conv2d(128, 64, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 256, 1, bias=False),
            nn.BatchNorm2d(256)
        )
        self.bypass = nn.Sequential(
            nn.Conv2d(128, 256, 1, bias=False),
            nn.BatchNorm2d(256)
        )

    def forward(self, x):
        return (self.direct(x) + self.bypass(x)).relu()

batch_size = 4
x = torch.rand(batch_size, 128, 16, 16) # тензор x в программе не менять

model_bn = Bottleneck()
model_bn.eval()

y = model_bn(x)
y.shape
```

```python
import torch
import torch.nn as nn

class Block_NN(nn.Module):
    def __init__(self):
        super().__init__()
        self.block1 = nn.Sequential(
            # Вход: (batch_size, 128, 16, 16)
            nn.Conv2d(128, 64, (1, 1), stride=1, padding=0, bias=False),  # (batch_size, 64, 16, 16)
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, (3, 3), stride=1, padding=1, bias=False),  # (batch_size, 64, 16, 16)
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 256, (1, 1), stride=1, padding=0, bias=False),  # (batch_size, 256, 16, 16)
            nn.BatchNorm2d(256),
        )

        self.block2 = nn.Sequential(
            # Вход: (batch_size, 128, 16, 16)
            nn.Conv2d(128, 256, (1, 1), stride=1, padding=0, bias=False),  # (batch_size, 256, 16, 16)
            nn.BatchNorm2d(256),
        )

    def forward(self, x):
        x = self.block1(x) + self.block2(x)
        x.relu_()
        return x

batch_size = 4
x = torch.rand(batch_size, 128, 16, 16) # тензор x в программе не менять

model_bn = Block_NN()
model_bn.eval()

y = model_bn(x)

y.shape
```
