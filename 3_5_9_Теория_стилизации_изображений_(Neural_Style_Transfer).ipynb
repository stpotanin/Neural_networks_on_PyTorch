````python
import torch

a, b, n = map(float, input().split()) # переменные a, b, n в программе не менять
# a, b, n = -2, 3, 10

n = int(n)
t_range = torch.linspace(a, b, n)
t_out = t_range.sigmoid()
t_pred = torch.argmax(t_out).item()

t_pred
````
