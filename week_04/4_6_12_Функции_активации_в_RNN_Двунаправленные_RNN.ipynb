```python
import torch
import torch.nn as nn

class Two_Inp(nn.Module):
    def __init__(self):
        super().__init__()
        self.inp_1 = nn.Sequential(nn.Linear(7, 12),
                                  nn.Tanh())
        self.inp_2 = nn.Sequential(nn.Linear(12, 12),
                                  nn.Sigmoid())
        self.out = nn.Sequential(nn.Linear(12, 32),
                                  nn.ReLU(),
                                  nn.Linear(32, 1))

    def forward(self, a, b):
        return self.out(self.inp_1(a) + self.inp_2(b))

batch_size=12
a = torch.rand(batch_size, 7) # тензоры a, b в программе не менять
b = torch.rand(batch_size, 12)

model = Two_Inp()
model.eval()
predict = model(a, b)
predict
```
