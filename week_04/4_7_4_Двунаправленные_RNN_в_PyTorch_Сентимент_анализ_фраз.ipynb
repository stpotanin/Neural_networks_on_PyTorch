```python
import torch
import torch.nn as nn

class Bidirect_Model(nn.Module):
    def __init__(self, d_size, hidden_size, out_size):
        super().__init__()
        self.rnn = nn.RNN(d_size, hidden_size, batch_first=True, bidirectional=True)
        self.linear = nn.Linear(hidden_size * 2, 5)

    def forward(self, x):
        _, h = self.rnn(x)
        h = torch.cat([h[0], h[1]], dim=1)
        return self.linear(h)

batch_size = 8
seq_length = 12
d_size = 32
x = torch.rand(batch_size, seq_length, d_size)

in_linear = 24               # из рисунка
out_size = 5                 # из рисунка
hidden_size = in_linear // 2 # т.к. bidirectional

model = Bidirect_Model(d_size, hidden_size, out_size)
model.eval()
predict = model(x)
predict
```
