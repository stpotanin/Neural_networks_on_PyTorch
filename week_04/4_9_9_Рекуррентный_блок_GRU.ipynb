```python
import torch
import torch.nn as nn

class GRU_Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.gru = nn.GRU(
            input_size=2,
            hidden_size=5,
            batch_first=True,
            bidirectional=True,
            bias=False
        )

        self.bn = nn.BatchNorm1d(num_features=10)
        self.linear = nn.Linear(10, 4)

    def forward(self, x):
        gru_out, _ = self.gru(x)  # (3, 17, 10)
        # Нужно переставить оси, чтобы num_features=10 стало второй размерностью.
        gru_out = gru_out.transpose(1, 2)   # (3, 10, 17)
        gru_norm = self.bn(gru_out)         # (3, 10, 17)
        gru_norm = gru_norm.transpose(1, 2) # (3, 17, 10) - обратно

        return self.linear(gru_norm)            # (3, 17, 4)

# тензор x в программе не менять
batch_size = 3
seq_length = 17
d_size = 2
x = torch.rand(batch_size, d_size)
# делаем тензор нужного формата: (batch_size, seq_length, in_features)
xx = torch.cat((x.unsqueeze(1), torch.zeros(batch_size, seq_length - 1, d_size)), dim=1)

model = GRU_Model()
model.eval()
predict = model(xx)
predict.shape   # (3, 17, 4)
```

```python
import torch
import torch.nn as nn

class GRU_Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.gru = nn.GRU(
            input_size=2,
            hidden_size=10,
            batch_first=True
        )

        self.bn = nn.BatchNorm1d(num_features=10)
        self.linear = nn.Linear(10, 4)

    def forward(self, x):
        gru_out, _ = self.gru(x)                # h0 будет нулевым по умолчанию
        gru_out = gru_out.transpose(1, 2)       # (batch, hidden, seq)
        gru_norm = self.bn(gru_out)
        gru_norm = gru_norm.transpose(1, 2)     # обратно: (batch, seq, hidden)
        out = self.linear(gru_norm)             # применяем linear к каждому timestep
        return out

# Тест
batch_size = 3
seq_length = 17
d_size = 2
x = torch.rand(batch_size, d_size)
xx = torch.cat((x.unsqueeze(1), torch.zeros(batch_size, seq_length - 1, d_size)), dim=1)

model = GRU_Model()
model.eval()
predict = model(xx)
print(predict.shape)  # (3, 17, 4)
```
