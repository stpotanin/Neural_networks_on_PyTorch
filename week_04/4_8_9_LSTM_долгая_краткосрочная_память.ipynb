```python
import torch
import torch.nn as nn

class OutputModule(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Linear(24, 5)

    def forward(self, x):
        # От LSTM x приходит в таком виде:
        lstm_out, (h_t, c_t) = x
        # Поскольку  линейный слой преобразует 24 признака в 5,
        # размер заготовки для выхода будет:
        out = torch.zeros(lstm_out.shape[0], lstm_out.shape[1], 5)

        for i in range(lstm_out.shape[1]):
            hidden_state = lstm_out[:, i, :]
            out[:, i, :] = self.layer(hidden_state)

        return out

# тензор x в программе не менять
batch_size = 4
seq_length = 64
in_features = 5
x = torch.rand(batch_size, seq_length, in_features)

model = nn.Sequential(
    nn.LSTM(
        input_size=5,   # Совпадает с in_features
        hidden_size=12, # Нужно для Linear(24, 5) с учётом bidirectional
        batch_first=True,
        bidirectional=True
    ),
    OutputModule()
)

model.eval()
out = model(x)
out.shape   # [4, 64, 5]
```

```python
# https://github.com/selfedu-rus/neuro-pytorch/blob/main/solves/4.8.7
import torch
import torch.nn as nn

# здесь объявляйте класс OutputModule
class OutputModule(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Linear(24, 5)

    def forward(self, x):
        batch_size = x[0].size(0)
        n = x[0].size(1)
        y = torch.empty(batch_size, n, self.layer.out_features)

        for i in range(n):
            y[:, i, :] = self.layer(x[0][:, i, :])
        return y


# тензор x в программе не менять
batch_size = 4
seq_length = 64
in_features = 5
x = torch.rand(batch_size, seq_length, in_features)

# здесь продолжайте программу
model = nn.Sequential(
    nn.LSTM(in_features, 12, batch_first=True, bidirectional=True),
    OutputModule(),
)

model.eval()
out = model(x)
out.shape, out  # [4, 64, 5]
```
