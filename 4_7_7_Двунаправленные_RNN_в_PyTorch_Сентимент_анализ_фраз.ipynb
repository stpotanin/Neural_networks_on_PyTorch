```python
import torch
import torch.nn as nn


class Combine(nn.Module):
    def __init__(self):
        super().__init__()
        # Many-to-One (bidirectional RNN)
        self.rnn_bi = nn.RNN(in_features, 9, batch_first=True, bidirectional=True)
        # One-to-Many (обычная RNN)
        self.rnn = nn.RNN(18, hidden_size, batch_first=True)
        self.out = nn.Linear(hidden_size, 3)

    def forward(self, x):
        # Many-to-One
        _, h = self.rnn_bi(x)                # h: [2, batch_size, hidden_size]
        h = torch.cat([h[0], h[1]], dim=-1)  # [batch_size, hidden_size*2]

        '''
        По условию: Начальные состояния всех векторов скрытых состояний равны нулю
        '''
        # Вектор только на первом (нулевом) шаге, остальные — нули:
        h_25 = torch.zeros(x.size(0), out_seq_len, h.size(1))
        h_25[:, 0, :] = h  # [batch_size, 25, hidden_size*2]

        # One-to-Many
        out_seq, _ = self.rnn(h_25)

        # Проекция в пространство выходных признаков
        x = self.out(out_seq)
        return x

# Параметры модели
batch_size = 2
seq_length = 12
in_features = 5
# Из рисунка
hidden_size = 32    # размер скрытого состояния (не умноженный!)
out_seq_len = 25
out_size = 3        # число выходных признаков на шаг

# тензор x в программе не менять
x = torch.rand(batch_size, seq_length, in_features)

model = Combine()

model.eval()
results = model(x)
results.shape, results   # [2, 25, 3]
```

```python
import torch
import torch.nn as nn

# здесь объявляйте класс модели
class MyModelRNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn_1 = nn.RNN(5, 9, batch_first=True, bidirectional=True)
        self.rnn_2 = nn.RNN(18, 32, batch_first=True)
        self.output = nn.Linear(32, 3)
        self.out_length = 25

    def forward(self, x):
        _, h = self.rnn_1(x)
        y = torch.cat([h[0], h[1]], dim=1)
        u = torch.zeros(y.size(0), self.out_length, y.size(1))
        u[:, 0, :] = y
        y, _ = self.rnn_2(u)

        n = y.size(1)
        out = torch.empty(y.size(0), n, self.output.out_features)

        for i in range(n):
            out[:, i, :] = self.output(y[:, i, :])
        return out


# тензор x в программе не менять
batch_size = 2
seq_length = 12
in_features = 5
# x = torch.rand(batch_size, seq_length, in_features)

# здесь продолжайте программу
model = MyModelRNN()

model.eval()
results = model(x)
results.shape, results   # [2, 25, 3]
```
