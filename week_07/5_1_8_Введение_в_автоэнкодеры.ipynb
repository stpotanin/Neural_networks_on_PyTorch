```python
# @title Без y (target) в датасете
import torch
import torch.nn as nn
import torch.utils.data as data
import torch.optim as optim

class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        # Кодер
        self.encoder = nn.Sequential(
            nn.Linear(8, 16),
            nn.ReLU(),
            nn.Linear(16, 4),
            nn.ReLU(),
            nn.Linear(4, 2),
            nn.Tanh()
        )

        # Декодер
        self.decoder = nn.Sequential(
            nn.Linear(2, 4),
            nn.ReLU(),
            nn.Linear(4, 8)
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        # тензор с выхода декодера, тензор с выхода кодера:
        return x_hat, z

total = 1000 # размер выборки
data_x = torch.rand(total, 8) # обучающие данные
train_data = data.DataLoader(data_x, batch_size=16, shuffle=True)

model = AutoEncoder()

optimizer = optim.RMSprop(params=model.parameters(), lr=0.001, weight_decay=0.0001)
loss_func = nn.MSELoss()
epochs = 5

model.train()

for _e in range(epochs):
    for x_train in train_data:
        predict, _ = model(x_train)
        loss = loss_func(predict, x_train)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

model.eval()

with torch.no_grad():
    predict, _ = model(data_x)

Q = loss_func(predict, data_x).item()
Q
```
