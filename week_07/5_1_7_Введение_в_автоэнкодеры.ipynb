```python
import torch
import torch.nn as nn

class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        # Кодер
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 8, (3, 3), stride=2, padding=1,
                      bias=False),  # Вход [3,16,16], Выход [8,8,8]
            nn.ReLU(),
            nn.BatchNorm2d(8),      # размер не меняется
            nn.Conv2d(8, 4, (3, 3), stride=2, padding=1,
                      bias=False),  # Вход [8,8,8], Выход [4,4,4]
            nn.ReLU(),
            nn.BatchNorm2d(4),      # размер не меняется
            nn.Flatten(),           # 4×4×4=64
            nn.Linear(64, 4, bias=True),
            nn.ReLU()
        )

        # Декодер
        self.decoder = nn.Sequential(
            nn.Linear(4, 64),
            nn.ELU(),
            nn.Unflatten(dim=1, unflattened_size=(4, 4, 4)),
            nn.ConvTranspose2d(4, 8, (2, 2), stride=2, padding=0, bias=True),
            nn.ELU(),
            nn.ConvTranspose2d(8, 1, (2, 2), stride=2, padding=0, bias=True)
        )

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        # тензор с выхода декодера, тензор с выхода кодера:
        return x_hat, z

h = torch.rand(4) # тензор h в программе не менять

model = AutoEncoder()

# # Для тестирования!
# st_model = model.state_dict()

model.load_state_dict(st_model)

# В режиме эксплуатации пропустите через ДЕКОДЕР полученной модели тензор:
model.eval()
out = model.decoder(h.unsqueeze(0))  # делаем батч размером 1

out.shape
```
