```python
import torch

a = {'мужчина': [-5, 4],
     'женщина': [-4, 1],
     'красный': [4, -2],
     'автомобиль': [6, 7],
     'машина': [10, 8]}

keys = ['машина', 'автомобиль', 'мужчина', 'женщина', 'красный']

emb_words = torch.IntTensor([a[i] for i in keys])

emb_words
```
