```python
import torch
import torch.nn as nn

class AutoEncoder_mini(nn.Module):
    def __init__(self, input_dim, output_dim, hidden_dim):
        super().__init__()
        # Извлекам угловой к-т в виде вектора (тензора)
        d = torch.tensor([7.0, 3.0])           # B - A
        d_unit = d / d.norm()                  # нормированный вектор

        # Точка A — начало прямой
        A = torch.tensor([0.0, 3.0])

        # Кодер
        self.encoder = nn.Linear(input_dim, hidden_dim, bias=True)
        self.encoder.weight.data = d_unit.unsqueeze(0)  # (1, 2)
        self.encoder.bias.data = torch.tensor([-d_unit @ A])  # скалярное произведение

        # Декодер
        self.decoder = nn.Linear(1, 2, bias=True)
        self.decoder.weight.data = d_unit.unsqueeze(1)  # (2, 1)
        self.decoder.bias.data = A  # чтобы t=0 → точка A

    def forward(self, x):
        z = self.encoder(x)
        x_hat = self.decoder(z)
        return x_hat

input_dim, output_dim, hidden_dim = 2, 2, 1
model_ae = AutoEncoder_mini(input_dim, output_dim, hidden_dim)

# Для проверки нужна точка ОБЯЗАТЕЛЬНО лежащая на прямой
x = torch.tensor([[14.0, 9.0]])  # (batch_size=1, features=2)
out = model_ae(x)
out
```
