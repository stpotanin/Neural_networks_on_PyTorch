```python
import torch
import torch.nn as nn


class TwoLayerModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(3, 2)
        self.layer2 = nn.Linear(2, 1)

    def forward(self, x):
        return self.layer2(self.layer1(x).sigmoid())


x = torch.rand(3) # тензор x в программе не менять

model = TwoLayerModel()
model.eval()
predict = model(x)
predict
```
