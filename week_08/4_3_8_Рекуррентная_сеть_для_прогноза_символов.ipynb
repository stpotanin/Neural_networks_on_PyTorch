```python
import torch

 # точки прямой
a = [[-2, 0], [0, 3]]

# задаём фиксироваанное значение для любого из трёх коэффициентов
w2 = -1

# по двум точкам находим коэффициент наклона и w1
k = (a[1][1] - a[0][1]) / (a[1][0] - a[0][0])
w1 = k

# используем любую точку для нахождения w0
w0 = -k * a[0][0] + a[0][1]

w = torch.tensor([w0, w1, w2], dtype=torch.float32)
w
```
