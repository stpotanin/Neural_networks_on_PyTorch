```python
import pandas as pd
import torch

srting = 'как работает нейронная сеть было показано на примере работы рекуррентной нейронной сети прогноза символов'
s = ['pad','pad'] + [word[0:3] for word in srting.split()] + ['pad','pad']  # список из основ слов в строке + паддинг=2

words = ['ней', 'сет', 'пок', 'при', 'раб', 'рек']
column_headers = ['как', 'раб', 'ней', 'сет', 'был', 'пок', 'на', 'при', 'рек', 'про', 'сим', 'pad']

# Создаем DataFrame, заполненный нулями
df = pd.DataFrame(0, index=words, columns=column_headers)

for word in words:               # по заголовкам строк
    for i in range(2, len(s)-2): # по словам в строке
        if word == s[i]:
            df.loc[word, s[i-2]] += 1
            df.loc[word, s[i-1]] += 1
            df.loc[word, s[i+1]] += 1
            df.loc[word, s[i+2]] += 1

emb_words = torch.IntTensor(df.iloc[:, :-1].values)
emb_words
```
